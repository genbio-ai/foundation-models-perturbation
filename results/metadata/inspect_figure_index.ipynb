```python
import json
import pandas as pd
import numpy as np
```

```python
with open('fig_index.json', 'r') as f:
    fig_index = json.load(f)
```

```python
total_models = 0
```

## Norman LFC

```python
# Names in CSV but not in a figure: norman_lfc.csv
df = pd.read_csv("/home/jovyan/pert-bench-fusion-paper/figures/norman_lfc.csv", index_col=0)
unique_names = df["name"].unique()
print(len(df["name"].unique()) , "models total in the CSV")
total_models += len(df["name"].unique())
print("Below are the ones present in the CSV but not in the figures")

unique_names = [x.replace("knn", "HEAD_TYPE").replace("lasso", "HEAD_TYPE") for x in unique_names]
unique_names = [str(x) for x in np.unique(unique_names) if x not in fig_index["norman_lfc"]]
unique_names
```
